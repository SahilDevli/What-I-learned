```python
# Stats function in python

import pandas as pd
import numpy as np
```

```python
df  = pd.DataFrame(np.random.rand(10,5))  # create random value table/data_freame of 10 rows and 5 columns.
print(df)
df.dtypes
```

```python
print(df.describe()) # all info about data column wise
```

```python
# can perform saparate according to the req.
# like - row or column wise , only on perticular column or row.

print(df.sum(axis=0)) # column wise sum (by default axis=0)
print(df.sum(axis=1)) # row wise sum

# simlilarly -> mean, max, min, var, std, corr.

print(df.mean(axis=0)) # mean of each column
print(df.mean(axis=1)) # mean of each row

print(df.max(axis=0)) # maximum value in each column
print(df.min()) # minimum value in each column

print(df.var()) # variance of each column
print(df.std()) # standard deviation of each column

print(df.corr()) # correlation matrix of all columns
```
