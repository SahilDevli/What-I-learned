```python
import pandas as pd
import numpy as np
```

```python
# numpy basics

arr = np.array([1,2,3,4,5])
print("1d array: ",arr)
arr2 = np.array([[1,2,3],[4,5,6]])
print("2d array: ",arr2)


# list vs numpy array
py_list = [1,2,3]   # [1,2,3,1,2,3]
print("Multiply Python list by 2: ",py_list * 2)

numpy_array = np.array([1,2,3])     # [2,6,8]
print("numpy_array mutiplication: ",numpy_array*2)
```
