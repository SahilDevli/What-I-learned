```python
# data frames -> python dic and csv
import pandas as pd
```

```python
dic = {
    'file_name': ['file1.c','file2.js','file3.jsx'],
    'file_size(kb)': [1000,2300,30000],
    'programming':['C programming','Java script','React JS']
}
```

```python
# data frames from dic
df = pd.DataFrame(dic)
print(df)

# data from from csv file
df2 = pd.read_csv("demo2.csv")
print(df2)

# convert data frame to csv file
df.to_csv("demo.csv")
```

```python
array = df.to_numpy() # convert dataset into numpy_array
print(array)

print(df.head(1))   # first 1 element of D.S.
print(df.tail(1))   # last 1 element of D.S.
```

```python
print(df.index)#information about indexing
print(df)
#change column name
df.columns = ['Name','Size']
print(df)
```

```python
# different functions in data frame
print('type : ',type(df))

#sorting
print(df.sort_index(axis=0,ascending=True))
print(df.sort_index(axis=0,ascending=False))

#copy and view of data set
view = df #this is view of data frame
copy = df.copy() # create a copy of data frame
```
