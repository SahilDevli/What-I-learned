```python
import pandas as pd
import numpy as np
```

```python
# importing another data set with numeric data and NaN values -> handle nan value

head = ["ID","Name","Status","Calories","Duration"]
df = pd.read_csv("prac.csv", names = head, header=0)
print(df)
```

```python
# df.replace(np.nan, 0, inplace=True) # chnage orignal data set
df.replace(np.nan, 0) # method with this parameter dont change orignam df and data set
```

```python
missing_value = df.isnull()
# to count number of true and false value in missing_value which set from .isnull() to identify NaN values
for column in head:
    print(column)
    print(missing_value[column].value_counts())
    print(" ")
```

# Cleaning this NaN vales as possible
## 1. Drop the Row or Column on the basis of data set
## 2. Replace missing values with mean (if less outliers), meadian (if outliers are more affective), mode (in some case like door in car).
## 3. leave missing data(not recomended)

```python
Cal_avg = df["Calories"].astype('float').mean()
# Cal_med = df["Calories"].astype('float').median()
# Cal_mod = df["Calories"].astype('float').mode()
df.replace({"Calories": Cal_avg}, inplace=True)
print(df)
```

```python
df = df.dropna(subset=["Status"])
df = df[df["Status"] != ""]

print(df)
```
